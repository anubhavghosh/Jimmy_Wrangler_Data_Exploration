# STEP 1: Importing first data set "Fortune_500_Headquarters_raw"
In this step, I have imported the first dataset and created the dataframe called "dataframe_1"

```python
import pandas as pd
dataframe_1 = pd.read_csv("Fortune_500_Headquarters_raw.csv")
dataframe_1.head()
```

# STEP 2: Importing second data set "fortune_1000_raw"
In this step, I have imported the second dataset and created the dataframe called "dataframe_2"
*It is important to check the encoding of the files.

```python
dataframe_2 = pd.read_csv("fortune1000_raw.csv")
dataframe_2.head()
```

# STEP 3: Cleaning the Dataframes
*Remove unneccessary columns from both the datasets.

```python
drop_cols_1 = ['X','Y','FID','OBJECTID','ADDRESS2','LATITUDE','LONGITUDE','SOURCE','PRC','COUNTYFIPS','COMMENTS','PROFIT']
dataframe_1.drop(drop_cols_1,axis=1, inplace=True)
```

```python
drop_cols_2 = ['Previous Rank','Assets ($M)','Employees','Latitude','Longitude','Years on Fortune 500 List','City']
dataframe_2.drop(drop_cols_2,axis=1,inplace=True)
```

# STEP 4: Merge the cleaned Dataframes

```python
dataframe_result=pd.merge(dataframe_1,dataframe_2, on='NAME')
dataframe_result
```
